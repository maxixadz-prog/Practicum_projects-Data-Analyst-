- Если вы нашли пропуски в данных с количеством проданных копий игры в том или ином регионе, их можно заменить на среднее значение в зависимости от названия платформы и года выхода игры.

```python
#Обрабатываем пропущенные значения - 1 удаляем минимальные пропуски 
games_min_omissions = games.dropna(subset=['name', 'genre', 'eu_sales', 'jp_sales'])

print(f"Удалено {len(games) - len(games_min_omissions)} строк с минимальными пропусками")
print(f"Осталось {len(games_min_omissions)} строк")
```

```python
# Обрабатываем пропуски 
import pandas as pd

# Сохраняем исходное количество строк
initial_quantity = len(games)

# 1. Удаляем минимальные пропуски
games = games.dropna(subset=['name', 'genre', 'eu_sales', 'jp_sales'])

# 2. Заполняем пропуски в продажах (средние по платформе+год)
for col in ['na_sales', 'eu_sales', 'jp_sales', 'other_sales']:
    avg_sales = games.groupby(['platform', 'year_of_release'])[col].mean()
    games[col] = games.apply(
        lambda row: avg_sales.get((row['platform'], row['year_of_release']), 0) 
        if pd.isna(row[col]) else row[col],
        axis=1
    )

# 3. Обрабатываем оценки
games['has_user_score'] = games['user_score'].notna().astype(int)
games['has_critic_score'] = games['critic_score'].notna().astype(int)
games['has_rating'] = games['rating'].notna().astype(int)
games['user_score'] = games['user_score'].fillna(-1)
games['critic_score'] = games['critic_score'].fillna(-1)
games['rating'] = games['rating'].fillna('No Rating')

# Вывод результатов
print(" РЕЗУЛЬТАТЫ ОБРАБОТКИ:")
print(f"• Удалено строк: {initial_quantity - len(games)}")
print(f"• Осталось строк: {len(games)}")
print(f"• Игр с оценками пользователей: {games['has_user_score'].sum()}")
print(f"• Игр с оценками критиков: {games['has_critic_score'].sum()}")
print(f"• Игр с рейтингом: {games['has_rating'].sum()}")
```

## 4. Категоризация данных
    
Проведите категоризацию данных:
- Разделите все игры по оценкам пользователей и выделите такие категории: высокая оценка (от 8 до 10 включительно), средняя оценка (от 3 до 8, не включая правую границу интервала) и низкая оценка (от 0 до 3, не включая правую границу интервала).

```python
# Создаем пустой список для категорий
game_groups = []

# Заполняем категории
for score in df_actual['user_score']:
    if score >= 8:
        game_groups.append("Высокая оценка")
    elif score >= 3:
        game_groups.append("Средняя оценка")
    elif score >= 0:
        game_groups.append("Низкая оценка")
    else:
        game_groups.append("Нет оценки")

# Добавляем в датафрейм
df_actual['user_score_category'] = game_groups

print(df_actual['user_score_category'].value_counts())
        
```

- Разделите все игры по оценкам критиков и выделите такие категории: высокая оценка (от 80 до 100 включительно), средняя оценка (от 30 до 80, не включая правую границу интервала) и низкая оценка (от 0 до 30, не включая правую границу интервала).

```python
# Создаем пустой список для категорий оценок критиков
critic_groups = []

# Заполняем категории
for score in df_actual['critic_score']:
    if score >= 80:
        critic_groups.append("Высокая оценка")
    elif score >= 30:
        critic_groups.append("Средняя оценка")
    elif score >= 0:
        critic_groups.append("Низкая оценка")
    else:
        critic_groups.append("Нет оценки")
        
#Добавляем в датафрейм
df_actual['critic_score_category'] = critic_groups        

# Смотрим сколько в каждой категории
print("Категории оценки критиков:")
print("Высокая оценка:",critic_groups.count("Высокая оценка"))
print("Средняя оценка:",critic_groups.count("Средняя оценка"))
print("Низкая оценка:",critic_groups.count("Низкая оценка"))
print("Нет оценки:",critic_groups.count("Нет оценки"))
```
